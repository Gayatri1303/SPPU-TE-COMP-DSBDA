```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("C:\\Users\\admin\\Downloads\\boston_housing.csv")
```

```python
df_x=df
df_y=df.price
print(df_y)    #here price is wht we want to achieve or target variable and rest others are used to predict the price,i.ie feature variable
```

```python
 def RemoveOutlier(df,var):
    Q1 = df[var].quantile(0.25)
    Q3 = df[var].quantile(0.75)
    IQR = Q3 - Q1
    high, low = Q3+1.5*IQR, Q1-1.5*IQR
    print("Highest allowed in variable:", var, high)
    print("lowest allowed in variable:", var, low)
    count = df[(df[var] > high) | (df[var] < low)][var].count()
    print('Total outliers in:',var,':',count)
    df = df[((df[var] >= low) & (df[var] <= high))]
    return df
```

```python
# Splitting the dataset into the Training set and Test set

xtrain, xtest, ytrain, ytest = train_test_split(df_x, df_y, test_size = 0.25, random_state = 13)
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
# Fitting linear regression to the Training set
from sklearn.linear_model import LinearRegression
model =  LinearRegression()
model.fit(xtrain, ytrain)
# Predicting the Test set results
ypred = model.predict(xtest)

print(ypred)
```

```python
print(ytest)
```

```python
print(np.mean((ypred-ytest)**2))
#The mean squared error function
```
