```python
from sklearn.preprocessing import StandardScaler

# Assuming df is your DataFrame and you want to scale the fourth column
data = df.iloc[:, 3].values  # Extracting the values of thefourth  column

# Reshape the data to a 2D array
data_reshaped = data.reshape(-1, 1)

# Create the StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the data
model = scaler.fit(data_reshaped)

# Transform the data
scaled_data = model.transform(data_reshaped)

print(scaled_data)
```
