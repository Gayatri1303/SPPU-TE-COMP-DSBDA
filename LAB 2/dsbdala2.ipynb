```python
import numpy as np
import pandas as pd
```

```python
data={
    'fname':['gaya','adv','sakh','sarthak','atharva','yash','mitisha','tanvi','bhumi','suhani','nitya'],
    'lname':['kuru','shinde','ingale','kar','garud','jangale','chawla','maideo','roy','desai','potdar'],
    'genedr':['F','M','F','M','M','M','F','F','F','F','F'],
    'DSBDA':[70,60,80,53,np.nan,90,88,np.nan,55,80,490],
    'WT':[89,90,87,np.nan,67,356,97,np.nan,64,84,45],
    'CC':[56,np.nan,54,65,98,np.nan,245,32,78,56,np.nan]
    
    
}
```

```python
df=pd.DataFrame(data)
df
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
mean_value = df['DSBDA'].mean()
df['DSBDA'].fillna(mean_value, inplace=True)
df
```

```python
#drop all rows with atleast one null value
new_df2 = df.dropna(axis=0 , how='any')
new_df2
```

```python
#dropping column with atleast one null value
new_df=df.dropna(axis=1,how='any')
new_df
```

```python
df
```

```python
#Val from next row
#bfill is used for taking values from next row
df['WT'].fillna(method='bfill' , inplace = True)
df
```

```python
#method pad for filling na values with val from prev row
df['CC'].fillna(method='pad',inplace=True)
df
```

```python
#for outliers
import seaborn as sns
import matplotlib.pyplot as plt


sns.boxplot(data=df,x=df['DSBDA'])
plt.show()
```

```python
sns.boxplot(data=df,x=df['WT'])
plt.show()
```

```python
sns.boxplot(data=df,x=df['CC'])
plt.show()
```

```python
df['DSBDA']=np.where((df['DSBDA'] > 100),df['DSBDA'].median(), df['DSBDA'])
df
```

```python
sns.boxplot(data=df,x=df['DSBDA'])
plt.show()
```

```python
#transformations

#log
df['CC']=np.log1p(df['CC'])
sns.boxplot(data=df,x=df['CC'])
plt.show()
```

```python
#scaling
from sklearn.preprocessing import StandardScaler

data = df.iloc[:, 5].values
reshaped_data = data.reshape(-1, 1)

scaler = StandardScaler()
model = scaler.fit(reshaped_data)
scaled_data = model.transform(reshaped_data)

print(scaled_data)
```

```python
sns.histplot(data=scaled_data)
plt.title('WT SCALED DATA')
plt.show()
```
